```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

# Import the data
df = pd.read_csv("insurance.csv")
# show the first five rows to visualize the data we are working
df.head()
```

---

Use the data in the relevant columns to determine how age affects insurance costs:
- Plot a scatter plot with age on the x-axis and charges on the y-axis.

```python
# scatterplot to see patterns in our observation
sns.scatterplot(data = df, x = 'age', y = 'charges')

#specfiy axis labels
plt.xlabel('Age')
plt.ylabel('Charges')

plt.show()
```

> Looking for the dataset is possible to see a linear relationship in the way the observations are distributed.

---

- Using linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on data.

```python
# Fit the model
x = df[['age']].values
y = df['charges'].values

insurance_model = LinearRegression()
insurance_model.fit(x, y)
```

```python
# Predict an unknown value
unk_x = [[36]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

sns.scatterplot(data = df, x = 'age', y = 'charges')

plt.plot(x_pred, y_pred, color = 'r')

plt.xlabel("Age")
plt.ylabel("Charges")

plt.show

print(f"Here is the insurance charge prediction for a {unk_x}-year-old individual:", insurance_model.predict(unk_x))
```

---

- Plot another scatter plot with the best-fit line.

```python
# Plot the data and model
y_pred = insurance_model.predict(x)

sns.scatterplot(data = df, x='age', y='charges')

plt.plot(x, y_pred, color = 'r')

plt.xlabel("Age")
plt.ylabel("Charges")

plt.show()
```
